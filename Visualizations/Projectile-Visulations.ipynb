```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import json_normalize
import json
```

```python
with open('../simulations/projectile-sim.json') as jsonfile:
    simdict = json.load(jsonfile)

df = pd.DataFrame(simdict["Snapshots"])
location_data = json_normalize(df["Location"])
location_data.rename(columns={'x':'Distance','y':'Height'},inplace=True)
location_data.drop(columns=['z'],inplace=True)
location_data.plot(x='Distance',y = 'Height')
```
