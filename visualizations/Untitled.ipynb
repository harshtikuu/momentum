```python
import pandas as pd
from pandas import json_normalize
import matplotlib.pyplot as plt
import json
```

```python
with open('spring-sims.json') as file:
    dictionary = json.load(file)
```

```python
df= pd.DataFrame(dictionary['Snapshots'])
```

```python
y = json_normalize(df['Acceleration'])['x']
x = df['Time']
```

```python
plt.plot(x,y)
```
